## Dataset 1: Netflix Movies and TV Shows

Este dataset contiene información sobre películas y series disponibles en la plataforma Netflix.
Cuenta con más de 8.000 registros y 12 variables, por lo que cumple con los requisitos de la consigna.

### Variables potencialmente interesantes

- **type**: permite distinguir entre películas y series de TV.
- **release_year**: útil para analizar la evolución del contenido a lo largo del tiempo.
- **country**: permite estudiar la distribución geográfica del contenido.
- **listed_in**: representa los géneros asociados a cada título.
- **duration**: permite comparar la duración de películas y la cantidad de temporadas en series.

### Contexto del dataset

Este conjunto de datos es útil para analizar la estrategia de contenidos de Netflix, identificar tendencias
en tipos de producciones y evaluar la diversidad del catálogo según país, género y año de lanzamiento.

## Dataset: Social Media and Mental Health

Este dataset contiene información sobre el uso de redes sociales y su relación con indicadores
de salud mental en distintos usuarios. Incluye datos demográficos, patrones de uso de plataformas
digitales y resultados de cuestionarios clínicos estandarizados.

El conjunto de datos cuenta con 8000 registros y 15 variables, cumpliendo con los requisitos
establecidos en la consigna.

### Variables potencialmente interesantes

- **Age**: edad del usuario, permite analizar diferencias generacionales en el uso de redes sociales.
- **Gender**: variable demográfica que permite estudiar posibles diferencias según género.
- **Primary_Platform**: plataforma principal utilizada por el usuario (por ejemplo, Instagram, TikTok, etc.).
- **Daily_Screen_Time_Hours**: cantidad de horas diarias frente a pantallas, variable clave para evaluar exposición.
- **Late_Night_Usage**: indica si el usuario utiliza redes sociales durante la noche.
- **Sleep_Duration_Hours**: horas de sueño diarias, relevante para analizar impacto en el descanso.
- **Social_Comparison_Trigger**: identifica si el usuario se ve afectado por comparaciones sociales.
- **GAD_7_Score** y **GAD_7_Severity**: puntaje y severidad de ansiedad según el cuestionario GAD-7.
- **PHQ_9_Score** y **PHQ_9_Severity**: puntaje y severidad de depresión según el cuestionario PHQ-9.

### Contexto del dataset

Este dataset resulta especialmente útil para analizar la relación entre el uso de redes sociales
y la salud mental. Permite estudiar cómo factores como el tiempo de exposición, el tipo de contenido
consumido y los hábitos nocturnos pueden estar asociados a niveles de ansiedad y depresión.

Además, ofrece una base sólida para análisis exploratorios, estudios correlacionales y futuros
modelos predictivos relacionados con bienestar digital y salud psicológica.

## Dataset: Spotify Tracks

Este dataset contiene información sobre canciones disponibles en la plataforma Spotify.
Incluye características de audio extraídas automáticamente, datos de popularidad y
clasificación por género musical.

El conjunto de datos cuenta con más de 100.000 registros y más de 20 variables,
cumpliendo ampliamente con los requisitos establecidos en la consigna.

### Variables potencialmente interesantes

- **popularity**: indica el nivel de popularidad de cada canción dentro de la plataforma.
- **danceability**: mide qué tan adecuada es una canción para bailar según sus características rítmicas.
- **energy**: representa la intensidad y actividad percibida de la canción.
- **valence**: describe el grado de positividad emocional transmitido por la canción.
- **tempo**: velocidad de la canción medida en pulsaciones por minuto (BPM).
- **loudness**: volumen promedio de la pista.
- **track_genre**: género musical asociado a la canción.
- **duration_ms**: duración de la canción en milisegundos.

### Contexto del dataset

Este dataset permite analizar patrones musicales y preferencias de los usuarios,
explorar relaciones entre características de audio y popularidad, y comparar géneros
musicales en función de variables cuantitativas.

Es especialmente útil para análisis exploratorios, visualización de datos y
modelos de recomendación musical.
