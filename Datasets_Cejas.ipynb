```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
url_mentalhealth= "https://raw.githubusercontent.com/meceugenia/datasets-/refs/heads/main/social_media_mental_health.csv"
mentalhealth=pd.read_csv(url_mentalhealth)

mentalhealth.head()
```

## Dataset: Social Media and Mental Health

Este dataset contiene información sobre el uso de redes sociales y su relación con indicadores
de salud mental en distintos usuarios. Incluye datos demográficos, patrones de uso de plataformas
digitales y resultados de cuestionarios clínicos estandarizados.



# Abstract

El presente trabajo analiza la relación entre el uso de redes sociales y distintos indicadores de salud mental a partir del dataset “Social Media and Mental Health”. Este conjunto de datos incluye 8000 registros y 15 variables que abarcan información demográfica, hábitos digitales y resultados de cuestionarios clínicos estandarizados como el GAD-7 (ansiedad) y PHQ-9 (depresión).

El objetivo principal es explorar si existe una asociación entre el tiempo de uso de redes sociales y los niveles de ansiedad y depresión reportados por los usuarios. En particular, se analizarán variables como la cantidad de horas diarias frente a pantallas, el uso nocturno de redes sociales, la duración del sueño y los puntajes obtenidos en las escalas clínicas.

La relevancia del estudio radica en el creciente debate sobre el impacto de las plataformas digitales en el bienestar psicológico, especialmente en poblaciones jóvenes. A través de visualizaciones univariadas, bivariadas y multivariadas, junto con resúmenes estadísticos, se buscará identificar patrones y tendencias que permitan responder a las preguntas de investigación planteadas.

Finalmente, se examinará la presencia de valores faltantes y se evaluará la consistencia de los datos, asegurando una base sólida para el análisis exploratorio.

### Variables potencialmente interesantes

- **Age**: edad del usuario, permite analizar diferencias generacionales en el uso de redes sociales.
- **Gender**: variable demográfica que permite estudiar posibles diferencias según género.
- **Primary_Platform**: plataforma principal utilizada por el usuario (por ejemplo, Instagram, TikTok, etc.).
- **Daily_Screen_Time_Hours**: cantidad de horas diarias frente a pantallas, variable clave para evaluar exposición.
- **Late_Night_Usage**: indica si el usuario utiliza redes sociales durante la noche.
- **Sleep_Duration_Hours**: horas de sueño diarias, relevante para analizar impacto en el descanso.
- **Social_Comparison_Trigger**: identifica si el usuario se ve afectado por comparaciones sociales.
- **GAD_7_Score** y **GAD_7_Severity**: puntaje y severidad de ansiedad según el cuestionario GAD-7.
- **PHQ_9_Score** y **PHQ_9_Severity**: puntaje y severidad de depresión según el cuestionario PHQ-9.

### Contexto del dataset

Este dataset resulta especialmente útil para analizar la relación entre el uso de redes sociales
y la salud mental. Permite estudiar cómo factores como el tiempo de exposición, el tipo de contenido
consumido y los hábitos nocturnos pueden estar asociados a niveles de ansiedad y depresión.

Además, ofrece una base sólida para análisis exploratorios, estudios correlacionales y futuros
modelos predictivos relacionados con bienestar digital y salud psicológica.

# Preguntas e Hipótesis

## Preguntas de investigación

1. ¿Existe relación entre la cantidad de horas diarias de uso de redes sociales y los niveles de ansiedad?
2. ¿El uso nocturno de redes sociales está asociado con menor duración del sueño?
3. ¿Las personas con mayor tiempo de pantalla presentan mayores niveles de depresión?

## Hipótesis

1. H1: A mayor cantidad de horas diarias frente a pantallas, mayor puntaje en ansiedad (GAD-7).
2. H2: El uso nocturno de redes sociales se asocia con menor duración del sueño.
3. H3: Existe una relación positiva entre tiempo de pantalla y puntaje de depresión (PHQ-9).

```python
plt.figure()
sns.boxplot(
    x="Late_Night_Usage",
    y="Sleep_Duration_Hours",
    data=mentalhealth
)
plt.title("Uso nocturno vs Horas de sueño")
plt.xlabel("Uso nocturno")
plt.ylabel("Horas de sueño")
plt.show()
```

## Interpretación : depresión promedio por género y uso nocturno

El gráfico de barras presenta el promedio del puntaje de depresión (PHQ-9) según el género de los participantes, diferenciando además entre quienes reportan uso nocturno de redes sociales y quienes no lo hacen.

Esta visualización permite analizar simultáneamente tres variables: género (variable demográfica), uso nocturno (variable conductual) y nivel de depresión (variable psicológica).

Al comparar las barras dentro de cada género, puede observarse si el uso nocturno se asocia con un mayor promedio de síntomas depresivos. Si las barras correspondientes a quienes utilizan redes sociales por la noche presentan valores más elevados, esto sugiere una posible asociación entre el uso nocturno y mayores niveles de depresión.

Asimismo, el gráfico permite identificar si existen diferencias generales en los niveles promedio de depresión entre géneros y evaluar si el efecto del uso nocturno es similar o diferente en cada grupo.

Este análisis contribuye a responder la hipótesis planteada sobre si los hábitos digitales nocturnos se relacionan con la salud mental, considerando además posibles diferencias demográficas.

```python
plt.figure()
sns.histplot(mentalhealth["Daily_Screen_Time_Hours"], kde=True)
plt.title("Distribución de horas diarias de pantalla")
plt.xlabel("Horas")
plt.ylabel("Frecuencia")
plt.show()
```

## Interpretación del Histograma: horas diarias de pantalla

El histograma muestra la distribución de la variable "Daily_Screen_Time_Hours", permitiendo observar cómo se distribuyen las horas de uso de redes sociales en la muestra.

Se puede identificar en qué rango horario se concentra la mayor cantidad de participantes y si existen valores extremos con niveles muy elevados de exposición. Este análisis univariado resulta fundamental para comprender el comportamiento general de la variable antes de analizar su relación con los niveles de ansiedad y depresión.

La forma de la distribución también permite detectar si los datos presentan sesgo o si se concentran en un intervalo específico de horas.

```python
plt.figure()
sns.boxplot(
    x="Late_Night_Usage",
    y="GAD_7_Score",
    data=mentalhealth
)
plt.title("Ansiedad según uso nocturno")
plt.xlabel("Uso nocturno")
plt.ylabel("Puntaje GAD-7")
plt.show()
```

## Interpretación del Boxplot: ansiedad según nivel de uso

El boxplot permite comparar la distribución de los puntajes de ansiedad (GAD-7) entre distintos niveles de uso de redes sociales.

Cada caja representa la mediana, los cuartiles y la dispersión de los puntajes dentro de cada categoría. Se puede observar si los niveles de ansiedad tienden a ser mayores en los grupos con mayor tiempo de exposición.

Si se identifican medianas más altas en los grupos de uso elevado, esto sugiere una posible asociación entre mayor tiempo frente a pantalla y mayores niveles de ansiedad.

Asimismo, la presencia de valores atípicos indica que existen individuos con puntajes particularmente altos o bajos en cada grupo.

```python
plt.figure()
sns.barplot(
    x="Gender",
    y="PHQ_9_Score",
    hue="Late_Night_Usage",
    data=mentalhealth
)
plt.title("Depresión promedio por género y uso nocturno")
plt.ylabel("Promedio PHQ-9")
plt.show()
```

```python
mentalhealth.groupby("Late_Night_Usage")[["GAD_7_Score","PHQ_9_Score"]].mean()
```

## Interpretación : depresión según género y uso nocturno

El gráfico de barras muestra el promedio de los puntajes de depresión (PHQ-9) según el género de los participantes, diferenciando además entre quienes reportan uso nocturno de redes sociales y quienes no lo hacen.

Este análisis permite observar simultáneamente tres variables: género, uso nocturno y nivel promedio de depresión.

Si dentro de cada género se observa que la barra correspondiente al uso nocturno presenta un promedio más elevado, esto sugiere que el uso de redes sociales en horario nocturno podría estar asociado con mayores niveles de síntomas depresivos.

Asimismo, el gráfico permite comparar si existen diferencias generales en los niveles de depresión entre géneros y evaluar si el impacto del uso nocturno varía entre ellos.

Este tipo de visualización resulta clave para identificar posibles patrones combinados entre factores conductuales y demográficos en relación con la salud mental.

```python
plt.figure()
sns.heatmap(
    mentalhealth[["Daily_Screen_Time_Hours","Sleep_Duration_Hours","GAD_7_Score","PHQ_9_Score"]].corr(),
    annot=True
)
plt.title("Matriz de Correlación")
plt.show()
```

## Interpretación del Heatmap: matriz de correlación

El heatmap representa la matriz de correlación entre variables numéricas relevantes del dataset, tales como horas de pantalla, duración del sueño, ansiedad (GAD-7) y depresión (PHQ-9).

Los colores indican la intensidad y dirección de la relación lineal entre variables. Las correlaciones positivas sugieren que ambas variables tienden a aumentar conjuntamente, mientras que las negativas indican que cuando una aumenta, la otra tiende a disminuir.

En particular, puede observarse una correlación positiva entre las horas diarias de pantalla y los puntajes de ansiedad y depresión, lo que sugiere una posible asociación entre mayor exposición y mayor malestar psicológico.

También se observa una relación negativa entre la duración del sueño y los niveles de ansiedad y depresión, lo que podría indicar que menos horas de descanso se asocian con mayor sintomatología.

Es importante aclarar que la correlación no implica causalidad, sino únicamente asociación estadística.
